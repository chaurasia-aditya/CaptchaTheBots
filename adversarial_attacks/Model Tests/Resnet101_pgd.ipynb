```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.callbacks import LambdaCallback
import tensorflow.keras.layers as L
from tensorflow.keras.datasets import mnist, cifar10
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
```

```python
import numpy as np
import matplotlib.pyplot as plt
from random import randint
from tqdm import tqdm
```

```python
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
# scaling the data

X_train_scaled = X_train/255

X_test_scaled = X_test/255
```

```python
nb_classes = 10
label_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog',
               'horse', 'ship', 'truck']
```

```python
from sklearn.linear_model import LogisticRegression

import numpy as np
from matplotlib import pyplot as plt

from art.estimators.classification import SklearnClassifier
from art.attacks.evasion import ProjectedGradientDescent
from art.utils import load_mnist

import warnings
warnings.simplefilter('ignore')

tf.compat.v1.disable_eager_execution()
```

```python
from keras.models import load_model
model = load_model('/kaggle/input/resnet101-keras/resnet101_keras_94.97.h5')
```

```python
from art.estimators.classification import KerasClassifier
mean_cifar10 = np.mean(X_train, axis=(0, 1, 2))
std_cifar10 = np.std(X_train, axis=(0, 1, 2))
classifier = KerasClassifier(
    clip_values=(0, 255),
    model=model,
    preprocessing=(mean_cifar10, std_cifar10)
)
```

```python
pgd = ProjectedGradientDescent(estimator=classifier, norm=np.inf, eps=.1, eps_step=0.01, max_iter=10,
                               targeted=False, num_random_init=0, batch_size=100)
```

```python
x_test_adv = pgd.generate(X_test_scaled)
```

```python
_, acc, *is_anything_else_being_returned = model.evaluate(x_test_adv, y_test, verbose=1)
print(f"Accuracy on Adversarial Examples: {acc*100}")
```

```python
predictions_adv = model.predict(x_test_adv)
predicted_labels_adv = np.argmax(predictions_adv, axis=1)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(predicted_labels_adv, y_test)
print(f'Accuracy: {accuracy * 100:.2f}%')
```
